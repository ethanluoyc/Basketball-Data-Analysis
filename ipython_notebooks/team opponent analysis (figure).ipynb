```python
import pandas as pd
import pylab as plt 
%matplotlib inline
```

```python
df= pd.read_csv('teamstats.csv')
df['WL']=False
df.ix[df['PTD']>0,'WL']=True     #the above steps determine the team wins/loses in a game
grouped = df.groupby('teamname')
mean = grouped.mean()            #this calculates the mean for all the teams involved in the season
teamlist =  mean.index.unique()
df = df.join(mean,on='teamname',rsuffix='_teammean')
df = df.join(mean,on='oppteamname',rsuffix='_oppmean')
df.to_csv('team_comparison.csv',index=False)
```

```python
plt.figure()
toplot = df.copy()
toplot['deviation']=toplot.TRBP-toplot.TRBP_mean
toplot['difference']=toplot.TRBP_mean-toplot.TRBP_oppmean
plt.scatter(toplot.difference,toplot.deviation)
plt.show()
print 'TRBP/TRBP'

plt.figure()
toplot = df.copy()
toplot['deviation']=toplot.FGP-toplot.FGP_mean
toplot['difference']=toplot.FGP_mean-toplot.FGP_oppmean
plt.scatter(toplot.difference,toplot.deviation)
plt.show()
print 'FGP/FGP'

plt.figure()
toplot = df.copy()
toplot['deviation']=toplot.ORBP-toplot.ORBP_mean
toplot['difference']=toplot.ORBP_mean-toplot.ORBP_oppmean
plt.scatter(toplot.difference,toplot.deviation)
plt.show()
print 'ORBP/ORBP'
```

```python
plt.figure()
toplot = df.copy()
toplot['deviation']=toplot.TRBP-toplot.TRBP_mean
toplot['difference']=toplot.WL_mean-toplot.WL_oppmean
plt.scatter(toplot.difference,toplot.deviation)
plt.show()
print 'TRBP/WL'

plt.figure()
toplot = df.copy()
# toplot['PTD']=toplot.TRBP-toplot.TRBP_mean
toplot['difference']=toplot.WL_mean-toplot.WL_oppmean
plt.scatter(toplot.difference,toplot.PTD)
plt.show()
print 'PTD/WL'
```

```python
df.plot(x='TRBP',y='PTD',kind='scatter')
df.plot(x='ORBP',y='PTD',kind='scatter')
df.plot(x='DRBP',y='PTD',kind='scatter')
```
